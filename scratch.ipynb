```python
from Bio import SeqIO

in_path  = "/Users/siegelmanfamily/Downloads/contig_1_Genome_R0_FLIPPED_Red_FASTA.fa"
out_path = "/Users/siegelmanfamily/Downloads/crackle_gene/GBK.gb"

# 1) Read all records from the FASTA
records = list(SeqIO.parse(in_path, "fasta"))

# 2) Add the required annotations to each record
for rec in records:
    rec.annotations["molecule_type"] = "DNA"       # required
    rec.annotations["topology"]      = "linear"    # usually linear or circular
    # you can also set rec.annotations["data_file_division"] = "PLN" or another code if you like

# 3) Write out as GenBank
count = SeqIO.write(records, out_path, "genbank")
print(f"Converted {count} record(s) to GenBank → {out_path}")
```
